### Adding titles,labels and legends to a chart

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = [5,10,15,20,25,30]

x2 = [1,2,3,4,5,6]
y2 = [2,4,8,16,32,64]

plt.xlabel("index")
plt.ylabel("count")
plt.title("My first line graph")

plt.plot(x1,y1,label="First Line")
plt.plot(x2,y2,label="Second Line")

plt.legend()
plt.show()
```

### Bar Charts

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = [5,10,15,20,25,30]

x2 = [1,2,3,4,5,6]
y2 = [2,4,8,16,32,64]

plt.xlabel("index")
plt.ylabel("count")
plt.title("My first bar graph")

plt.barh(x1,y1,label="Time 5",color="r")
plt.show()

plt.barh(x2,y2,label="Power 2",color="g")
plt.show()
```

## Stacked Bar Charts

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5,6]
y1 = [5,10,15,20,25,30]

x2 = [1,2,3,4,5,6]
y2 = [2,4,8,16,32,64]

plt.figure(figsize=(10,5))

plt.xlabel("index")
plt.ylabel("count")
plt.title("My first stacked bar graph")

plt.bar(x1,y1,label="Times 5")
plt.bar(x2,y2,label="Power 2")

plt.legend()
plt.show()
```

### Pie Charts

```python
sales = [10000,20000,30000,40000]
regions = ['North','South','East','West']

# user defined colors for slices
colors = ['r','g','b','y']

plt.pie(sales,labels=regions, colors=colors)
plt.show()
```

#### Pie Charts - Advanced Usage

```python
sales = [10000,20000,30000,40000]
regions = ['North','South','East','West']

# user defined colors for slices
colors = ['r','g','b','y']

plt.pie(sales,
        labels=regions,
        colors = colors,
       startangle=45,
       shadow=True,
       autopct = '%1.0f%%',
       explode=(0,0.6,0,0))
```

### Python Figure

```python
sales = [10000,20000,30000,40000]
regions = ['North','South','East','West']

# user defined colors for slices
colors = ['r','g','b','y']

plt.figure(figsize=(5,5))
plt.pie(sales,
        labels=regions,
        colors = colors,
       startangle=90,
       shadow=True,
       autopct = '%1.1f%%',
       explode=(0,0.4,0,0))
```

### Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# simple data to display in various forms
x = np.linspace(1,100,num=100)
y = np.random.normal(mu,sigma,100)

plt.close('all')

f,ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Simple Plot')

plt.show()
```

### Two subplots

```python
plt.close('all')

f,axarr = plt.subplots(2)
axarr[0].plot(x,y)
axarr[0].set_title('Two subplots')

axarr[1].hist(y)

plt.show()
```

### Two subplots, unpacking the axes array immediately

```python
plt.close('all')

f,(ax1,ax2) = plt.subplots(1,2,sharey=True)
ax1.plot(x,y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x,y)

plt.show()
```

### Two subplots sharing both x/y axesm

```python
plt.close('all')

f,(ax1,ax2) = plt.subplots(1,2,sharey=True)

ax1.plot(x,y)
ax1.set_title('Sharing Y Axis')
ax2.scatter(x,y)

plt.show()
```

### Three subplots sharing both x/y axesm

```python
plt.close('all')

f,(ax1,ax2,ax3) = plt.subplots(3,sharex=True,sharey=True)

ax1.plot(x,y)
ax1.set_title('Sharing both exes')

ax2.scatter(x,y)

ax3.scatter(x,2*y**2-1,color='r')

f.subplots_adjust(hspace=0)
plt.show()
```

### Subplots as a 2X2 grid

```python
fig, axs = plt.subplots(2, 2, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})
(ax1, ax2), (ax3, ax4) = axs


fig.suptitle('Sharing x per column, y per row')
axs[0][0].plot(x, y)
axs[0][1].plot(x, y**2, 'tab:orange')
axs[1][0].plot(x + 1, -y, 'tab:green')
axs[1][1].plot(x + 2, -y**2, 'tab:red')

for ax in axs.flat:
    ax.label_outer()
```
