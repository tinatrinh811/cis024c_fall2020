```python
alice_word_counts = {}
```

```python
for word in alice_contents_clean.split():
    
    temp_word = word.lower()
    
    if alice_word_counts.get(temp_word) == None:
        alice_word_counts[temp_word] = 1
    else:
        alice_word_counts[temp_word] = alice_word_counts[temp_word] + 1
```

### Write to a file

```python
employees = [
    ["1", 'Bob', "Physics"],
    ["2", 'Sarah', 'Computer Science'],
    ["3", ' John', 'Chemistry']
]
```

```python
employees
```

```python
employee_file = open('employee.txt','w')
employee_file
```

```python
?open
```

```python
for employee in employees:
    employee_file.write(",".join(employee))
    employee_file.write("\n")
```

```python
employee_file.close()
```

```python
## Reading the data back in

employee_list = []
employee_file = open('employee.txt','r')

count = 0
for line in employee_file.readlines():
    
    employee_list.append(line.strip().split(','))
    
    count = count + 1
```

```python
employee_list
```
